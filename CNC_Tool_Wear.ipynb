```
import os
import pandas as pd

# Folder containing CSV files
csv_folder = "/content/drive/MyDrive/CNC Experiments Files"  # Change this path

# List all CSV files in the folder
csv_files = [f for f in os.listdir(csv_folder) if f.endswith('.csv')]

# Create an empty DataFrame to store merged data
merged_df = pd.DataFrame()

# Loop through each CSV file and append to the DataFrame
for file in csv_files:
    file_path = os.path.join(csv_folder, file)
    df = pd.read_csv(file_path)  # Read CSV
    merged_df = pd.concat([merged_df, df], ignore_index=True)  # Merge

# Save the merged file
output_file = os.path.join(csv_folder, "merged_output.csv")
merged_df.to_csv(output_file, index=False)

print(f"Merged {len(csv_files)} CSV files into {output_file}")
```
